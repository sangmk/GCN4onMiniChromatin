```python
from scipy.special import comb
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(4,2))
n = np.array([i for i in range(9+1)])
plt.scatter(n, comb(9, n)*3**(9-n)/4**9)
plt.semilogy()
plt.ylabel('Probability')
plt.xlabel('Number of same base pairs.')
plt.title('When random sequence resembles ATGACTCAT')
```

对于长度为3070的chromatin来说，有3070-9=3061个长度为9的sequence（和zelin的计数方式一样，overlap的也算作两个）。这种情况下，重复4个base pair及以上的期望是：

```python
prob_atLeast4 = sum([comb(9, n)*3**(9-n)/4**9 for n in range(4,10)])
for N in [4,5,6,7]:
    exp_number = 3061 * sum([comb(9, n)*3**(9-n)/4**9 for n in range(N,10)])
    print(f"Expectation of sites w/ at least {N} same base pairs:",int(exp_number))
```
